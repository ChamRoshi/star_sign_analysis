```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_excel("preped_target.xlsx") 
features = pd.read_csv("features.csv")
```

```python
features = features[[x for x in features.columns if x not in ["april" ,
"november"   ,
"december",
"september" ,
"may"     ,   
"august"   ,  
"june"   ,   
"october"   ,
"july"  ,   
"january",
"march",  
"february" ]]]
```

```python
df.head()
```

```python
features.head()
```

```python
features.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, df["sign"], test_size=0.33, random_state=42)
```

## Multinomial NB

```python
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
1/12
```

```python
print(classification_report(y_test, clf.predict(X_test)))
```

```python
coefs = pd.DataFrame(clf.coef_, columns = features.columns, index = clf.classes_)
```

```python
for col in coefs.index: 
    print(col)
    print((coefs.T[col] - coefs.T.mean(axis = 1)).sort_values(ascending = False).head(30))
    print()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
r = RandomForestClassifier()
r.fit(X_train, y_train)
```

```python
print(classification_report(y_test, r.predict(X_test)))
```
